```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
from pprint import pprint
import random

wv_keys = wv.index_to_key # type: ignore

print(f'Il database contiene {len(wv_keys)} token, ognuno rappresentato come un vettore di {len(wv[0])} numeri.')

num = 10
print(f'\n{num} token scelti casualmente:')
pprint([wv_keys[random.randint(0, len(wv_keys))] for i in range(num)])

print(f'\nPer esempio, il primo token, {wv_keys[0]}, è rappresentato come:')
pprint(wv[0])
```

```python
pairs = [
    ('book', 'text'),
    ('book', 'pen'),
    ('Italy', 'Paris'),
    ('Italy', 'Paris'),
    ('Italy', 'Paris'),
    ('car', 'airplane'),
    ('car', 'cereal'),
    ('car', 'communism'),
]
for w1, w2 in pairs:
    print('%r\t%r\t%.2f' % (w1, w2, wv.similarity(w1, w2))) # type: ignore
```

```python
pprint(wv.most_similar(positive=['book'], topn=10)) # type: ignore
```
