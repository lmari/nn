```python
import torch
import matplotlib.pyplot as plt

device = torch.device("cuda:0")
dtype = torch.float

# funzione scalare: x.x
x = torch.arange(4, requires_grad=True, device=device, dtype=dtype)
y = torch.dot(x, x)
y.backward()
print(x.grad)

# funzione vettoriale: x*x
y0 = torch.ones(4, device=device, dtype=dtype)
x2 = torch.arange(4, requires_grad=True, device=device, dtype=dtype)
y2 = x2 * x2
y2.backward(gradient=y0)
print(x2.grad)

x3 = torch.arange(0, 10, 0.1, requires_grad=True, device=device, dtype=dtype)
y3 = torch.sin(x3)
y3.backward(gradient=torch.ones(len(x3), device=device, dtype=dtype))
plt.plot(x3.detach().cpu().numpy(), y3.detach().cpu().numpy())
plt.plot(x3.detach().cpu().numpy(), x3.grad.detach().cpu().numpy())
plt.show()
```
